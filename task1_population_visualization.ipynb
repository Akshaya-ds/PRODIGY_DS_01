```python
import zipfile
import os
import pandas as pd

# Extract the uploaded ZIP file
with zipfile.ZipFile('API_SP.POP.TOTL_DS2_en_csv_v2_38144.zip', 'r') as zip_ref:
    zip_ref.extractall('population_data')

# Load the main population CSV
df = pd.read_csv('population_data/API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv', skiprows=4)
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter 2022 data
df_2022 = df[['Country Name', '2022']].dropna()
df_2022.columns = ['Country', 'Population_2022']

# Plot Histogram
plt.figure(figsize=(10,6))
sns.histplot(df_2022['Population_2022'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Country Populations in 2022')
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Filter data for the year 2023
df_2023 = df[['Country Name', '2023']].dropna()
df_2023.columns = ['Country', 'Population_2023']

# Get top 10 most populated countries in 2023
top10_2023 = df_2023.sort_values(by='Population_2023', ascending=False).head(10)

# Plot bar chart
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.barplot(data=top10_2023, x='Country', y='Population_2023', palette='viridis')
plt.title('Top 10 Most Populated Countries in 2023')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Filter data for the year 2019
df_2019 = df[['Country Name', '2019']].dropna()
df_2019.columns = ['Country', 'Population_2019']

# Get top 10 most populated countries in 2019
top10_2019 = df_2019.sort_values(by='Population_2019', ascending=False).head(10)

# Plot bar chart
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.barplot(data=top10_2019, x='Country', y='Population_2019', palette='magma')
plt.title('Top 10 Most Populated Countries in 2019')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Remove rows with missing 2019 population values
df_hist = df[['Country Name', '2019']].dropna()
df_hist.columns = ['Country', 'Population_2019']

# Plot histogram
plt.figure(figsize=(10, 6))
plt.hist(df_hist['Population_2019'], bins=20, color='teal', edgecolor='black')
plt.title('Distribution of Country Populations in 2019')
plt.xlabel('Population Ranges')
plt.ylabel('Number of Countries')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
